```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from bokeh.plotting import figure, output_file, show
%matplotlib inline
```

```python
output_file(r"N:\Planning - New File Structure\Performance\Dashboards\dashboard.html")
```

```python
stats=pd.read_excel(r"\\VRIDE-FS2\Development\Planning - New File Structure\Performance\RidershipAutomation.xlsx")
stats=pd.DataFrame(stats)
```

```python
ridership=stats.loc[stats['Stat']=='VRH']
ridership.columns=ridership.columns.astype(str)
ridership.set_index(['Route','Area','Service'],inplace=True)
ridership.head()
```

```python
ridership_Q4_18=ridership.loc[:,'2018-07-01 00:00:00':'2018-09-01 00:00:00']
ridership_Q4_18=pd.DataFrame(ridership_Q4_18)
ridership_Q4_18
```

```python
ridership_Q4_18_route=ridership_Q4_18.groupby('Route').sum()
ridership_Q4_18_route['Total']=ridership_Q4_18_route.sum(axis=1)
ridership_Q4_18_route.drop(ridership_Q4_18_route.columns.difference(['Route','Total']),1,inplace=True)
ridership_Q4_18_route=ridership_Q4_18_route.sort_values('Total',ascending=False)
ridership_Q4_18_route.head()
```

```python
ridership_Q4_18_area=ridership_Q4_18.groupby(['Service','Area']).aggregate(sum)
ridership_Q4_18_area['Total']=ridership_Q4_18_area.sum(axis=1)
ridership_Q4_18_area.drop(ridership_Q4_18_area.columns.difference(['Area','Total']),1,inplace=True)
ridership_Q4_18_area=ridership_Q4_18_area.sort_values('Total',ascending=False)
ridership_Q4_18_area
```

```python
ridership_Q4_18_service=ridership_Q4_18.groupby('Service').aggregate(sum)
ridership_Q4_18_service['Total']=ridership_Q4_18_service.sum(axis=1)
ridership_Q4_18_service.drop(ridership_Q4_18_service.columns.difference(['Service','Area','Total']),1,inplace=True)
ridership_Q4_18_service=ridership_Q4_18_service.sort_values('Total',ascending=False)
ridership_Q4_18_service
```

```python
fig, ax=plt.subplots()
ax.axis('equal')
servicepie, _= ax.pie(ridership_Q4_18_service, radius=2)
plt.setp( servicepie,width=.5,edgecolor='white')
areapie, _ = ax.pie(ridership_Q4_18_area, radius=1.5)
plt.setp( areapie, width=.4, edgecolor='white')
routepie, _ = ax.pie(ridership_Q4_18_route, radius=1.4-.3)
plt.setp( routepie, width=.3, edgecolor='white')
plt.margins(0,0)
```
